```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data_files/random_data/data_music.csv')
```

```python
data = data.drop(['Unnamed: 0'],axis = 1)
```

```python
X = data.iloc[:,1:]
Y = data.iloc[:,0]
```

```python
X.head()
```

```python
Y.head()
```

```python
X.columns
```

```python
X.head()
```

```python
X = pd.concat([X,pd.get_dummies(X['relative_time'],prefix='relative_time')],axis=1)
X.drop(['relative_time'],axis=1, inplace=True)
```

```python
X.columns
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.datasets import make_classification
X1, Y1 = make_classification(n_samples=5000,
                             n_features=60, 
                             n_redundant=0,n_classes=10,
                             n_informative=60)
```

```python
X_synthetic = pd.DataFrame(data=X1,columns=X.columns)
```

```python
Y_synthetic = pd.DataFrame(data=Y1,columns=['music'])
```

```python
data_cloth_synthetic = pd.concat([Y_synthetic,X_synthetic],axis=1)
```

```python
data_cloth_synthetic.info()
```

```python
data_cloth_synthetic.to_csv('../data_files/synthetic_data/data_music_synthetic.csv')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_synthetic.values, Y_synthetic.values, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix 
knn = KNeighborsClassifier(n_neighbors = 7).fit(X_train, y_train) 
  
# accuracy on X_test 
accuracy = knn.score(X_test, y_test) 
print(accuracy)

# creating a confusion matrix 
knn_predictions = knn.predict(X_test)  
cm_knn = confusion_matrix(y_test, knn_predictions) 
```

```python
import pickle
pickle.dump(knn, open("../model_weights/pkl_weights/model_music.pkl","wb"))
```
